```python
import pandas as pd
import numpy as np
```

```python
def load_data_from_github(url: str) -> pd.DataFrame:
    df = pd.read_csv(url)
    return df
```

```python
def get_max_close(symbol):
    return symbol['Close'].max()
```

```python
if __name__ == '__main__':
    AMD = load_data_from_github(
        'https://raw.githubusercontent.com/huseinzol05/Stock-Prediction-Models/refs/heads/master/dataset/AMD.csv'
    )
    GOOG = load_data_from_github(
        'https://raw.githubusercontent.com/huseinzol05/Stock-Prediction-Models/refs/heads/master/dataset/AMD.csv'
    )

    print(f'Max Close AAPL: {AMD["Close"].max()}')
    print(f'Max Close IBM: {GOOG["Close"].max()}')
```

```python

start_date = '2018-05-23'
end_date = '2019-05-23'

dates = pd.date_range(start=start_date, end=end_date)
```

```python
TWTR = load_data_from_github(
    'https://raw.githubusercontent.com/huseinzol05/Stock-Prediction-Models/refs/heads/master/dataset/TWTR.csv'
)
```

```python
base = TWTR[['Date', 'TWTR']].copy()
base = base.loc[:, ~base.columns.duplicated()].copy()
base = base.join(AMD[['AMD']],how='left')
base = base.join(GOOG[['GOOG']],how='left')
```

```python
base[['TWTR','AMD']]
```

```python
import matplotlib.pyplot as plt

def plot(df,title='Close Price History'):
    plt.plot(df)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close')
    plt.show()
```

```python
plot(df= base['AMD'],title='AMD')
plot(df= base['GOOG'],title='GOOG')
```
